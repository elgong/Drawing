```python

##################  保存成pdf  放入word格式更清晰
# Plot Bars

from pylab import *
from matplotlib.backends.backend_pdf import PdfPages

import matplotlib.pyplot as plt
from matplotlib import pyplot
import random
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
%matplotlib inline

plt.rcParams['savefig.dpi'] = 300 #图片像素
# 图片大小
plt.figure(figsize=(8,6))



names = ["1288","1682", "2563","3346","5002"]
x = [1288,1682,2563,3346,5002]
y_1 = [0.917,0.932,0.943,0.959,0.971]
y_2  = [0.979,0.986,0.994,0.994,0.995]

plt.subplots_adjust(left=1, bottom=None, right=2, top=None,
                wspace=20, hspace=2)


plt.xlim(1000,5200)  # 限定横轴的范围
plt.ylim(0.90, )  # 限定纵轴的范围


plt.plot(x, y_1, marker='s', mec='b', color = 'blue',mfc='b',label='不采用非线性前馈法',markersize=10,lw=2)
plt.plot(x, y_2, marker='o', mec='r', mfc='r',color = 'red', ms=10,label='采用非线性前馈法',markersize=10,lw=2)

### loc  图例位置
plt.legend(loc=4,fontsize=22)  # 让图例生效

    
plt.grid(b=True, 
         color='r', 
         linestyle='--', 
         linewidth=1, 
         alpha=0.3, 
        # axis='xy', 
         which="major")

plt.margins(0)
plt.subplots_adjust(bottom=0.10)
plt.xlabel('输入功率(W)',fontsize=22) #X轴标签
plt.ylabel("功率因数",fontsize=22) #Y轴标签

pyplot.xticks([1000,2000,3000,4000,5000],size = 22)
pyplot.yticks([0.92,0.94,0.96,0.98,1.0],size = 22)
plt.tight_layout()


#plt.title("A simple plot") #标题
plt.show()
#plt.savefig('D:\\f1.png')
```
